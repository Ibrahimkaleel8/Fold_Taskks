Text Preprocessing

```python
import nltk
nltk.download('punkt')
```

```python
import nltk
from nltk.tokenize import word_tokenize

def preprocess_text(text):
  tokens = word_tokenize(text)
  return [token.upper() for token in tokens]
```

Extract intents from transaction descriptions

```python
def extract_intents(description, intents):
  description=preprocess_text(description)
  transaction_intents = []

  for i in intents:
    tag = i["tag"]
    patterns = i["patterns"]
    for pattern in patterns:
      if any(word in description for word in preprocess_text(pattern)):
        transaction_intents.append(tag)
        break

  return transaction_intents
```

Recommend tags and get relevant tasks

```python
from collections import Counter

def recommend_tags(description, intents):
  transaction_intents=extract_intents(description, intents)
  tag_counter = Counter(transaction_intents)
  top_tags=tag_counter.most_common(3)

  return [tag for tag, _ in top_tags]
```

Print the rcommended tags for transaction

```python
import string
sample_Transaction = input()
Transaction=''.join([i.replace('/',' ') for i in sample_Transaction])
recommended_tags = recommend_tags(Transaction,intents)
print("Recommended Tags", recommended_tags)
```
